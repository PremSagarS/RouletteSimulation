# Consider that a player can only do a single type of bet one round where bettype is an independant type bet

```python
for bettype in ["COLOURS", "LOWHIGH", "FIVEBET", "EVENODD", "COLUMN", "STREET", "DOZENS", "STRAIGHTUP"]:
    # Getting expected values for different number of bets
    minRange = BETCOUNTSRANGES[bettype][0]
    maxRange = BETCOUNTSRANGES[bettype][1]
    events = {}
    for numOfBets in range(minRange + 1, maxRange + 1):
        winRate = numOfBets * (len(BETTYPES[bettype][1]) / 38)
        winEvent = ODDS[bettype] * 100
        loseRate = 1 - winRate
        loseEvent = 0
        events[numOfBets] = {'winRate': winRate, 'loseRate': loseRate, 'winEvent': winEvent, 'loseEvent': loseEvent}

    # Plotting
    xpoints = []
    ypoints = []
    for numOfBets in events:
        xpoints.append(numOfBets)
        expectedValue = events[numOfBets]['winRate'] * events[numOfBets]['winEvent'] + events[numOfBets]['loseRate'] * events[numOfBets]['loseEvent']
        expectedValue -= numOfBets * 100
        ypoints.append(expectedValue)
    plt.plot(xpoints, ypoints, marker='o')
    plt.title(bettype)
    plt.show()
```
